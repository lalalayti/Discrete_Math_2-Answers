## ITEM NO. 1 - Kruskal's Algorithm

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G"])
edgelist = [("A", "C"), ("A", "I"), ("C", "B"), ("C", "E"), ("C", "D"), ("B", "E"), ("I", "E"), ("E", "F"), ("F", "G"), ("G", "E")]
G.add_edges_from(edgelist)

highlighted_edges = [("A", "C"), ("A", "I"), ("C", "B"), ("C", "E"), ("C", "D"), ("E", "F"), ("G", "E")]

plt.figure(figsize=(10, 8))
pos = nx.planar_layout(G)
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={("A", "C"): 10, ("A", "I"): 5, ("C", "B"): 20, ("C", "E"): 15, ("C", "D"): 30, ("B", "E"): 25, ("I", "E"): 40, ("E", "F"): 35, ("F", "G"): 50, ("G", "E"): 45},
    font_color='red'
)
nx.draw(G, pos=pos, with_labels=True)
nx.draw_networkx_edges(G, pos, edgelist=highlighted_edges, edge_color='green', width=3) 
nx.draw(
    G, pos, edge_color='grey', width=1, linewidths=1,
    node_size=1000, node_color='yellow', alpha=0.9,
    with_labels=True
)

plt.show()
```

## ITEM NO. 2 - Prim's Algorithm

```python
import matplotlib.pyplot as plt
import networkx as nx
import random

random.seed(55)

G = nx.Graph()

G.add_nodes_from(["A", "B", "C", "D", "E", "F", "I", "G"])

edgelist = [("A", "C", 10), ("A", "I", 5), ("C", "B", 20), ("C", "E", 15),
            ("C", "D", 30), ("B", "E", 25), ("I", "E", 40), ("E", "F", 35),
            ("F", "G", 50), ("G", "E", 45)]
G.add_weighted_edges_from(edgelist)

pos = nx.planar_layout(G)  

plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_color='yellow', node_size=1000, alpha=0.9, edge_color='grey', width=1, linewidths=1)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): d['weight'] for u, v, d in G.edges(data=True)}, font_color='red')

mst = nx.minimum_spanning_tree(G, algorithm='prim')

nx.draw(mst, pos, with_labels=True, node_color='yellow', node_size=1000, edge_color='green', width=3, alpha=0.6)

plt.show()
```

## ITEM NO. 3 - Djitraka's Shortest Path Algorithm

```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the graph
G = nx.Graph()
G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G"])
edgelist = [("A", "C"), ("A", "I"), ("C", "B"), ("C", "E"), ("C", "D"), ("B", "E"), ("I", "E"), ("E", "F"), ("F", "G"), ("G", "E")]
G.add_edges_from(edgelist)

# Calculate shortest paths using Dijkstra's algorithm
shortest_paths = nx.single_source_dijkstra_path(G, "A")

# Extract the shortest path from "A" to "G"
shortest_path_nodes = shortest_paths["G"]

# Highlight the shortest path edges in green
highlighted_edges = [(shortest_path_nodes[i], shortest_path_nodes[i + 1]) for i in range(len(shortest_path_nodes) - 1)]

# Draw the graph
plt.figure(figsize=(10, 8))
pos = nx.planar_layout(G)
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={("A", "C"): 10, ("A", "I"): 5, ("C", "B"): 20, ("C", "E"): 15, ("C", "D"): 30, ("B", "E"): 25, ("I", "E"): 40, ("E", "F"): 35, ("F", "G"): 50, ("G", "E"): 45},
    font_color='red'
)
nx.draw(G, pos=pos, with_labels=True)
nx.draw_networkx_edges(G, pos, edgelist=highlighted_edges, edge_color='green', width=3)  # Highlight shortest path edges in green
nx.draw(
    G, pos, edge_color='GREY', width=1, linewidths=1,
    node_size=1000, node_color='yellow', alpha=0.9,
    with_labels=True
)

plt.show()
```
