## 1. Graphs


```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

vertices = ['A', 'B', 'C']
G.add_nodes_from(vertices)

edges = [('A', 'B'), ('B', 'C')]
G.add_edges_from(edges)

G.add_edge('C', 'C')

pos = nx.spring_layout(G)  
nx.draw(G, pos, with_labels=True, node_color='yellow', edge_color='gray', node_size=1000, font_size=15)

edge_labels = {('A', 'B'): 'e1', ('B', 'C'): 'e2', ('C', 'C'): 'e3'}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

plt.show()
```

This example illustrates a simple graph which consists of a finite, nonempty set of vertices $V = \{A, B, C\}$ and a finite set of edges $E = \{e_1, e_2, e_3\}$. Each edge is associated with a set of one or two vertices, fulfilling the edge-endpoint function requirement: $e_1 $ connects vertices $A$ and $B$, $e_2$ connects vertices $B$ and $C$, and $e_3$ is a loop at vertex $C$. This structure satisfies the definition of a graph, as it includes vertices and edges with defined endpoints. 

## 2. Digraphs

```python
import networkx as nx
import matplotlib.pyplot as plt

DG = nx.DiGraph()

vertices = ['A', 'B', 'C', 'D']
DG.add_nodes_from(vertices)

edges = [('A', 'B'), ('A', 'C'), ('B', 'C'), ('C', 'D'), ('D', 'A')]
DG.add_edges_from(edges)

pos = nx.spring_layout(DG)  
nx.draw(DG, pos, with_labels=True, node_color='lightpink', edge_color='gray', node_size=1000, font_size=15, arrowsize=20)

plt.show()
```

This directed graph consists of four vertices labeled $ A $, $ B $, $ C $, and $ D $, with five directed edges. The edges are directed as follows: from $ A $ to $B$, from $A$ to $C $, from $B$ to $C$, from $C$ to $D$, and from $D$ to $A$. This setup creates a cycle where each edge has a direction, showing how one vertex connects to another in a specific order.

## 3. Bipartite Graphs

```python
import networkx as nx
import matplotlib.pyplot as plt

U = range(1, 4)
V = range(4, 8)

B = nx.Graph()

B.add_nodes_from(U, bipartite=0)
B.add_nodes_from(V, bipartite=1)

for u in U:
    for v in V:
        B.add_edge(u, v)

pos = nx.bipartite_layout(B, U)
nx.draw(B, pos, with_labels=True, node_color=['lightgreen' if node in U else 'violet' for node in B.nodes()], node_size=700, font_size=16, font_color='black', edge_color='gray')
plt.show()
```

## 4. Simple Graph

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

nodes = ['A', 'B', 'C', 'D', 'E']
G.add_nodes_from(nodes)

edges = [('A', 'B'), ('A', 'C'), ('B', 'D'), ('C', 'D'), ('D', 'E')]
G.add_edges_from(edges)

pos = nx.spring_layout(G)  
nx.draw(G, pos, with_labels=True, node_color='lightpink', node_size=700, font_size=16, font_color='black', edge_color='gray')
plt.show()
```

This is an example of a simple graph where each edge connects two different vertices (no loops).
There is at most one edge between any pair of vertices (no parallel edges).

## 5. Complete Graphs

```python
import networkx as nx
import matplotlib.pyplot as plt

n = 5
K_n = nx.complete_graph(n)

mapping = {i: chr(65 + i) for i in range(n)}
K_n = nx.relabel_nodes(K_n, mapping)

pos = nx.spring_layout(K_n)
nx.draw(K_n, pos, with_labels=True, node_color='orange', node_size=700, font_size=16, font_color='black', edge_color='gray')
plt.show()
```

This is an example of a complete graph $K_n$ with 𝑛 vertices where every pair of distinct vertices is connected by a unique edge. This means that in 
$K_n$, each vertex is connected to every other vertex exactly once.

## 6. Subgraphs

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('C', 'D'), ('D', 'E'), ('C', 'E')])

H_nodes = ['A', 'B', 'C', 'D']
H = G.subgraph(H_nodes)

pos = nx.spring_layout(G)

plt.figure(figsize=(12, 6))

plt.subplot(121)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, font_size=16, font_color='black', edge_color='gray')
plt.title('Graph G')

nx.draw_networkx_edges(G, pos, edgelist=H.edges(), edge_color='blue', width=2)
nx.draw_networkx_nodes(G, pos, nodelist=H_nodes, node_color='lightgreen', node_size=700)
nx.draw_networkx_labels(G, pos, font_size=16, font_color='black')

plt.subplot(122)
nx.draw(H, pos, with_labels=True, node_color='lightgreen', node_size=700, font_size=16, font_color='black', edge_color='blue')
plt.title('Subgraph H')

plt.show()
```

The subgraph $ H $ of the graph $ G $ is formed by the vertices $ \{A, B, C, D\} $ and the edges $ \{\{A, B\}, \{A, C\}, \{B, D\}, \{C, D\}\} $, which are all subsets of the vertices and edges of $ G $. While $ G $ includes an additional vertex $ E $ and the edges $\{D, E\}$ and $\{C, E\}$, these are excluded in $ H $. As a subgraph, $ H $ preserves the connectivity and relationships between the included vertices exactly as they exist in $ G $, effectively demonstrating the concept of a subgraph by maintaining a coherent structure derived from a portion of the original graph.

## 7. Connected Graph

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

nodes = ['A', 'B', 'C', 'D', 'E']
G.add_nodes_from(nodes)

edges = [('A', 'B'), ('A', 'C'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('B', 'E')]
G.add_edges_from(edges)

pos = nx.spring_layout(G)  
nx.draw(G, pos, with_labels=True, node_color='green', node_size=700, font_size=16, font_color='white', edge_color='gray')
plt.show()
```

The graph $ G $ is a simple connected graph with 5 vertices $ A, B, C, D, E $ and 6 edges. It forms a basic geometric shape where each vertex is connected to at least one other vertex, ensuring that there is a path between every pair of vertices. This connectivity is visually evident in the graph as every vertex can be reached from every other vertex through a series of edges. This property of connectedness makes it a fundamental example in graph theory, illustrating the concept of a connected graph.

## 8. Handshake Theorem

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from(['A', 'B', 'C', 'D'])

G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('C', 'D'), ('B', 'C')])

pos = {
    'A': (0, 1),
    'B': (-1, 0),
    'C': (1, 0),
    'D': (0, -1)
}

nx.draw(G, pos, with_labels=True, node_color='purple', node_size=1000, font_size=20, font_color='white', edge_color='gray')

plt.show()
```

In this example graph, we have four vertices labeled $A, B, C$, and $D$, connected by three edges. Vertex $A$ is connected to vertices $B and C,$ vertex $B$ is connected to vertices $A, C$, and $D$, vertex $C$ is connected to vertices $A, B$, and $D$, and vertex $D$ is connected to vertices $B$ and $C$. Applying the Handshake Theorem, which states that the sum of the degrees of all vertices in an undirected graph is equal to twice the number of edges, we observe that the sum of the degrees of vertices $A, B, C,$ and $D$ is $10$. Dividing this sum by $2$, we find that the graph indeed has $5$ edges. This confirms the Handshake Theorem, as the sum of the degrees of the vertices is twice the number of edges.

## 9. Travelling in a Graph

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F'])

G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'E'), ('C', 'F'), ('D', 'E'), ('E', 'F')])

pos = {
    'A': (0, 1),
    'B': (-1, 0.5),
    'C': (1, 0.5),
    'D': (-1.5, 0),
    'E': (0, 0),
    'F': (1.5, 0)
}

nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=1000, font_size=20, font_color='black', edge_color='gray')

plt.show()
```

In this graph:
- Vertex A is connected to vertices B and C.
- Vertex B is connected to vertices A, C, and D.
- Vertex C is connected to vertices A, B, E, and F.
- Vertex D is connected to vertices B and E.
- Vertex E is connected to vertices D, B, C, and F.
- Vertex F is connected to vertices C and E.

For example, to travel from vertex A to vertex F, one possible path would be: A -> B -> C -> E -> F. This path follows the adjacency edges between the vertices.

## 10. Euler's Circuit

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from(['A', 'B', 'C', 'D'])

G.add_edges_from([('A', 'B'), ('A', 'D'), ('B', 'C'), ('D', 'C')])

pos = {
    'A': (0, 1),
    'B': (1, 1),
    'C': (1, 0),
    'D': (0, 0)
}

nx.draw(G, pos, with_labels=True, node_color='green', node_size=1000, font_size=20, font_color='white', edge_color='gray')

plt.show()
```

In this graph:
- Vertex $A$ is connected to vertices $B$ and $D$. 
- Vertex $B$ is connected to vertices $A$ and $C$.
- Vertex $C$ is connected to vertices $B$ and $D$.
- Vertex $D$ is connected to vertices $A$ and $C$. <br>
This graph has an Eulerian circuit because it is a connected graph where every vertex has an even degree. Starting from any vertex, you can traverse each edge exactly once and return to the starting vertex, forming a closed loop or circuit. In this example, an Eulerian circuit could start from vertex $A$, then move to $B, C, D$, and finally return to $A$.

## 11. Euler's trail

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F'])

G.add_edges_from([
    ('A', 'B'), ('B', 'C'),
    ('A', 'D'), ('A', 'E'),
    ('B', 'E'), ('B', 'F'),
    ('C', 'F'),
    ('D', 'E'), ('E', 'F')
])

pos = {
    'A': (0, 2),
    'B': (1, 2),
    'C': (2, 2),
    'D': (0, 1),
    'E': (1, 1),
    'F': (2, 1)
}

nx.draw(G, pos, with_labels=True, node_color='brown', node_size=1000, font_size=20, font_color='white', edge_color='gray')

plt.show()
```

This graph contains an Eulerian trail, which is a path that visits every edge exactly once. An example of an Eulerian trail in this graph could be: $A -> B -> C -> F -> E -> D -> A.$

- Vertex A is connected to vertices B and D.
- Vertex B is connected to vertices A, C, D, and E.
- Vertex C is connected to vertices B and F.
- Vertex D is connected to vertices A, B, and E.
- Vertex E is connected to vertices B, D, and F.
- Vertex F is connected to vertices C and E.

## 12. Hamiltonian Circuit 

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

nodes = ['A', 'B', 'C', 'D']
G.add_nodes_from(nodes)

edges = [('A', 'B'), ('A', 'C'), ('B', 'D'), ('C', 'D')]
G.add_edges_from(edges)

pos = {'A': (0, 1), 'B': (-1, 0), 'C': (1, 0), 'D': (0, -1)}
nx.draw(G, pos, with_labels=True, node_color='purple', node_size=1000, font_size=16, font_color='yellow', edge_color='gray')
plt.show()
```

- Vertex A is connected to vertices B and C.
- Vertex B is connected to vertices A and D.
- Vertex C is connected to vertices A and D.
- Vertex D is connected to vertices B and C. <br><br>
A Hamiltonian circuit in this graph would be: $A -> B -> D -> C -> A.$ This sequence visits each vertex exactly once, except for the first and the last, which are both vertex A, forming a closed loop. This circuit traverses through every vertex of the graph, fulfilling the requirements of a Hamiltonian circuit.
